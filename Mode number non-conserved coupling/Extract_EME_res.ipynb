read S matrix data from EME simulation

```python
import pandas as pd
import numpy as np
import glob
import re
from Functions import *
```

```python
# len_prop_arr = np.linspace(200,1500,651)

# RDL
filename = "../data/Vertical_EME/"
csv_filename = "./results/EME_res_w_RDL.csv"
len_prop_arr = np.linspace(100,3500,681)

# no RDL
foldername = "../data/Vertical_EME - RDL - long prop/"
csv_filename = "./results/EME_res_wo_RDL.csv"
len_prop_arr = np.linspace(100,10000,991)     # no-RDL
```

```python
# 获取所有匹配的文件
filename_list = glob.glob(foldername + 'len_taper_RDL-*.txt')
filename_list
```

```python
# 创建一个空的DataFrame用于存储所有数据
df_data = pd.DataFrame()
file_list = []
# 遍历每个文件
for filename in filename_list:
    # 使用正则表达式提取xxx数字部分
    match = re.search(foldername+r'len_taper_RDL-(\d+)\.txt', filename)
    len_taper = int(match.group(1))
    file_list.append((len_taper,filename))

file_list.sort(key=lambda file: file[0])

df_data['len_prop'] = len_prop_arr
for len_taper, file in file_list:
    data = pd.read_csv(file, header=None, names=[len_taper])
    if df_data.empty:
        df_data = data
    else:
        df_data = pd.concat([df_data, data], axis=1)

# 显示结果
print(df_data)
df_data.to_csv(csv_filename,index=False)
```

Plot field profile at different positions of the RDL layer

```python
foldername = "../data/EME profile/"
end_profile_filename = "EME_field_profile_end.txt"
field = np.loadtxt(foldername + end_profile_filename, delimiter=None, skiprows=1)
field = np.matrix_transpose(field)
```

```python
EME_y_min       = -10
EME_y_max       = 10
EME_z_min       = -10
EME_z_max       = 20
```

```python
xticks = np.arange(0,np.shape(field)[1],100)
xtickslabel = EME_y_min + (EME_y_max-EME_y_min)/np.shape(field)[1] * xticks
xtickslabel = np.array(["{:.1f}".format(label) for label in xtickslabel])
yticks = np.arange(0,np.shape(field)[0],50)
ytickslabel = EME_z_min + (EME_z_max-EME_z_min)/np.shape(field)[0] * yticks
ytickslabel = np.array(["{:.1f}".format(label) for label in ytickslabel])
ytickslabel
```

```python
start_profile_filename = "EME_field_profile_start.txt"
field = np.loadtxt(foldername + start_profile_filename, delimiter=None, skiprows=1)
field = np.matrix_transpose(field)

param_dict = {"figsize"        : [8,6],
              "component_list" : ["Re"],
              "title"          : "Field profile in RDL region (left end)",
              "fontsize"       : 6,
              "shrink"         : 0.3,
              "pad"            : 0.1,
              "xlabel"         : r"X($\mu m$)",
              "ylabel"         : r"Y($\mu m$)",
              "xticks"         : xticks,
              "yticks"         : yticks,
              "xtickslabel"    : xtickslabel ,
              "ytickslabel"    : ytickslabel}
Plot_field_profile(field, "Ex",**param_dict)
```
